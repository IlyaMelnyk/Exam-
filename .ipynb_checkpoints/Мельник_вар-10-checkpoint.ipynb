```python
# Splitting the dataset into the Training set and Test set
X = df.iloc[:, :-1]
y = df['Profit']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

# Multiple Linear Regression

```python
#!pip install statsmodels
```

```python
from sklearn.linear_model import LinearRegression
# Fitting Multiple Regression
mr = LinearRegression().fit(X_train, y_train)

# Coefficient of determination R^2
mr.score(X_train, y_train).round(3), mr.score(X_test, y_test).round(3)
```

```python
# Visualising the Test set results
plt.scatter(X_test['RD_Spend'], y_test, color = 'red')
plt.plot(X_test['RD_Spend'], mr.predict(X_test), 'bo')
plt.title('Profit vs RD_Spend (Test set)')
plt.xlabel('RD_Spend')
plt.ylabel('Profit')
plt.show()
```

```python
# Visualising the Training set results
plt.scatter(X_train['RD_Spend'], y_train, color = 'red')
plt.plot(X_train['RD_Spend'], mr.predict(X_train), 'bo')
plt.title('Profit vs RD_Spend (Training set)')
plt.xlabel('RD_Spend')
plt.ylabel('Profit')
plt.show()
```

Модель множ. регрессии получилась хорошей, коэф. детерминации высокий
